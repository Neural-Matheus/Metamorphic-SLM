```python
from pathlib import Path
from typing import List, Tuple
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import os
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores.faiss import FAISS

MODEL_NAME = model_name
FAISS_DIR = Path("../ExecutionModule/faiss")
CSV_PATH = Path("./retrieval_summary.csv")
OUTPUT_CSV = Path("./euclidean_fairness_scores.csv")
TOP_K = 4

def load_index(path: Path) -> Tuple[FAISS, HuggingFaceEmbeddings]:
    embed = HuggingFaceEmbeddings(model_name=MODEL_NAME)
    index = FAISS.load_local(str(path), embeddings=embed, allow_dangerous_deserialization=True)
    return index, embed


def retrieve(index: FAISS, q: str, k: int) -> List[str]:
    return [doc.page_content for doc in index.similarity_search(q, k=k)]


def embed_texts(model: HuggingFaceEmbeddings, texts: List[str]) -> np.ndarray:
    vecs = np.vstack(model.embed_documents(texts)).astype(np.float32, copy=False)
    norms = np.linalg.norm(vecs, axis=1, keepdims=True)
    norms = np.where(norms == 0.0, 1.0, norms)
    return vecs / norms


def dist_matrix(a: np.ndarray, b: np.ndarray) -> np.ndarray:
    a2 = (a ** 2).sum(1, keepdims=True)
    b2 = (b ** 2).sum(1, keepdims=True).T
    return np.sqrt(np.maximum(a2 + b2 - 2 * a @ b.T, 0.0))


def mean_topk(mat: np.ndarray, k: int) -> float:
    m = mat.copy()
    sel = []
    for _ in range(k):
        idx = np.unravel_index(np.argmin(m, axis=None), m.shape)
        sel.append(m[idx])
        m[idx[0], :] = np.inf
        m[:, idx[1]] = np.inf
    return float(np.mean(sel))


def hamming(a: List[int], b: List[int]) -> float:
    return sum(x != y for x, y in zip(a, b)) / len(a)


def highlight_pairs(mat: np.ndarray, k: int) -> Tuple[np.ndarray, np.ndarray]:
    m = mat.copy()
    rows, cols = [], []
    for _ in range(k):
        idx = np.unravel_index(np.argmin(m, axis=None), m.shape)
        rows.append(idx[0])
        cols.append(idx[1])
        m[idx[0], :] = np.inf
        m[:, idx[1]] = np.inf
    return np.array(rows), np.array(cols)


def plot_heat(mat: np.ndarray, pid: str, out: Path) -> None:
    fig, ax = plt.subplots(figsize=(4, 4))
    im = ax.imshow(mat, cmap="viridis", origin="upper")
    ax.set_xticks(range(mat.shape[1]))
    ax.set_yticks(range(mat.shape[0]))
    ax.set_xlabel("PERTURBED")
    ax.set_ylabel("NORMAL")
    ax.set_title(pid)
    r, c = highlight_pairs(mat, TOP_K)
    for rr, cc in zip(r, c):
        ax.add_patch(Circle((cc, rr), 0.4, facecolor="none", edgecolor="white", linewidth=2))
    fig.colorbar(im, ax=ax)
    fig.tight_layout()
    fig.savefig(out)
    plt.close(fig)


def main() -> None:
    index, embedder = load_index(FAISS_DIR)
    df = pd.read_csv(CSV_PATH)
    parse = lambda s: list(map(int, s.split()))
    results = []
    for i, row in df.iterrows():
        labels_a = parse(row["RetrievedLabelsNormalText"])
        labels_b = parse(row["RetrievedLabelsPerturbedText"])
        texts_a = retrieve(index, row["InputText"], TOP_K)
        texts_b = retrieve(index, row["PerturbedText"], TOP_K)
        emb_a = embed_texts(embedder, texts_a)
        emb_b = embed_texts(embedder, texts_b)
        mat = dist_matrix(emb_a, emb_b)
        mean_d = mean_topk(mat, TOP_K)
        ham = hamming(labels_a, labels_b)
        score = mean_d + ham
        results.append(dict(InputTextID=row["InputTextID"], MeanDist=mean_d, Hamming=ham, Score=score))
        print(f"{row['InputTextID']} -> MeanDist={mean_d:.4f}  Hamming={ham:.4f}  Score={score:.4f}")
        if i < PLOT_LIMIT:
            pid = f"{row['InputTextID']}_{i}"
            plot_heat(mat, pid, PLOT_DIR / f"{pid}.png")
    pd.DataFrame(results).to_csv(OUTPUT_CSV, index=False)
    print(f"CSV saved to {OUTPUT_CSV}")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('./euclidean_fairness_scores.csv')
TOP_K = 4


score_col = 'Score' if 'Score' in df.columns else df.columns[-1]

def categorize(score):
    if score == 0:
        return '0 (Perfectly stable)'
    elif score <= 0.63:
        return '0–0.63 (High robustness)'
    elif score <= 1.31:
        return '0.63–1.31 (Moderate robustness)'
    else:
        return '>1.31 (Low robustness)'

df['Range'] = df[score_col].apply(categorize)

order = ['0 (Perfectly stable)', '0–0.5 (High robustness)', '0.5–1.5 (Moderate robustness)', '>1.5 (Low robustness)']
counts = df['Range'].value_counts().reindex(order).fillna(0)

fig, ax = plt.subplots()
ax.bar(range(len(order)), counts.values)
ax.set_xticks(range(len(order)))
ax.set_xticklabels(order, rotation=45, ha='right')
ax.set_xlabel('Score Range')
ax.set_ylabel('Number of queries')
ax.set_title('Robustness Distribution (Score Euclidean)')
fig.tight_layout()
plt.show()
```

```python
from pathlib import Path
from typing import Tuple
import pandas as pd
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores.faiss import FAISS

FAISS_DIR   = Path("../ExecutionModule/faiss")
SUMMARY_CSV = Path("./retrieval_summary.csv")
SCORES_CSV  = Path("./euclidean_fairness_scores.csv")

MODEL_NAME  = "sentence-transformers/all-MiniLM-L6-v2"
TOP_K       = 4   # k no similarity_search
SAMPLE_N    = 2   # number of examples per category

# ────── load data ──────
summary = pd.read_csv(SUMMARY_CSV)
scores  = pd.read_csv(SCORES_CSV)
summary["Score"] = scores["Score"]

# ────── Calculate quartiles ──────
q1 = summary["Score"].quantile(0.25)
q3 = summary["Score"].quantile(0.75)
print(f"Q1 (25%): {q1:.4f}")
print(f"Q3 (75%): {q3:.4f}\n")

# ────── Categorization ──────
def categorize(score: float) -> str:
    if score == 0.0:
        return "Zero"
    if score <= q1:
        return "Low"
    if score <= q3:
        return "Medium"
    return "High"

summary["Category"] = summary["Score"].apply(categorize)

print("Count per category:")
print(summary["Category"].value_counts(), "\n")

# ────── Load FAISS retriever ──────
embed = HuggingFaceEmbeddings(model_name=MODEL_NAME)
index = FAISS.load_local(str(FAISS_DIR), embeddings=embed, allow_dangerous_deserialization=True)

# ────── Exibit docs ──────
def show_docs(query: str):
    docs = index.similarity_search(query, k=TOP_K)
    for doc in docs:
        print(f"  • Text: {doc.page_content}\n    Toxicity: {doc.metadata.get('label')}")
    print()

# ────── Samples ──────
for cat in ["Zero", "Low", "Medium", "High"]:
    subset = summary[summary["Category"] == cat]
    if subset.empty:
        continue

    print(f"\n=== Categoria: {cat} (n={len(subset)}) ===")
    sample = subset.sample(n=min(SAMPLE_N, len(subset)), random_state=42)
    for _, row in sample.iterrows():
        print(f"\n-- InputTextID: {row['InputTextID']}  |  Score: {row['Score']:.4f} --\n")
        print("  NORMAL PROMPT:")
        show_docs(row["InputText"])
        print("  PERTURBED PROMPT:")
        show_docs(row["PerturbedText"])
        print("-" * 80)
```
