```python
import re
from typing import List, Dict
import pandas as pd
from langchain_community.vectorstores.faiss import FAISS
from langchain.embeddings import OpenAIEmbeddings

# -- Constants --------------------------------------------------------------

# Suffix to remove in each query
_CLEAN_SUFFIX_PATTERN = re.compile(
    r"\s*Analyse the sentiment of this text as positive, negative or neutral\.\s*$"
)

# -- Helper Functions -------------------------------------------------------

def clean_query(text: str) -> str:
    """
    Remove "Analyse the sentiment of this text as positive, negative or neutral."
    , if it exists
    """
    return _CLEAN_SUFFIX_PATTERN.sub("", text)


def load_and_clean_queries(csv_path: str) -> pd.DataFrame:
    """
    Read only the relevant colums of the original sheet and clean queries
    Returns a DataFrame with columns:
      - InputTextID
      - InputText        (clean)
      - PerturbedText    (clean)
    """
    df = pd.read_csv(csv_path, usecols=["InputTextID", "InputText", "PerturbedText"])
    df["InputText"]     = df["InputText"].apply(clean_query)
    df["PerturbedText"] = df["PerturbedText"].apply(clean_query)
    return df


def retrieve_labels(
    index: FAISS,
    query: str,
    k: int
) -> List[int]:
    """
    Execute similarity_search and extract only the labels (toxicity 0/1)
    of the k more similar documents 
    """
    docs = index.similarity_search(query, k=k)
    return [int(doc.metadata.get("label", 0)) for doc in docs]


# -- Main Orchestration -----------------------------------------------------

def run_retrieval_and_summarize(
    csv_path: str,
    faiss_index_path: str,
    output_path: str,
    top_k: int = 5
) -> None:
    # 1) read and clean queries
    df_queries = load_and_clean_queries(csv_path)

    # 2) load FAISS index
    faiss_index = FAISS.load_local(
        faiss_index_path,
        embeddings=embedding_model,
        allow_dangerous_deserialization=True
    )

    # 3) iterate over each query and collect results
    records: List[Dict] = []
    for _, row in df_queries.iterrows():
        nid       = row["InputTextID"]
        normal    = row["InputText"]
        perturbed = row["PerturbedText"]

        labels_norm   = retrieve_labels(faiss_index, normal,    k=top_k)
        labels_pert   = retrieve_labels(faiss_index, perturbed, k=top_k)

        records.append({
            "InputTextID":                  nid,
            "InputText":                    normal,
            "PerturbedText":                perturbed,
            "RetrievedLabelsNormalText":    " ".join(map(str, labels_norm)),
            "RetrievedLabelsPerturbedText": " ".join(map(str, labels_pert)),
            "ToxicCountNormalText":         sum(labels_norm),
            "ToxicCountPerturbedText":      sum(labels_pert),
        })

    # 4) Mount and save DataFrame
    summary_df = pd.DataFrame(records)
    summary_df.to_csv(output_path, index=False)


if __name__ == "__main__":
    CSV_INPUT_PATH   = "../ExecutionModule/CreatedFunctions_Fairness_Iteration0_sentiment analysis_llama.csv"
    FAISS_INDEX_PATH = "../ExecutionModule/faiss"
    CSV_OUTPUT_PATH  = "./retrieval_summary.csv"
    TOP_K            = 4

    run_retrieval_and_summarize(
        csv_path=CSV_INPUT_PATH,
        faiss_index_path=FAISS_INDEX_PATH,
        output_path=CSV_OUTPUT_PATH,
        top_k=TOP_K
    )
```

```python
import os
import pickle
import numpy as np
from typing import Dict

# Adjust imports according to your environment
try:
    from langchain_community.vectorstores.faiss import FAISS
except ModuleNotFoundError:
    from langchain.vectorstores import FAISS

import faiss

class FaissEmbeddingExtractor:
    def __init__(self, faiss_store: FAISS):
        """
        Initialize extractor with a LangChain FAISS store.
        """
        self._faiss_index: faiss.Index = faiss_store.index
        self._id_map: Dict[int, str] = faiss_store.index_to_docstore_id

    def get_all_embeddings(self) -> Dict[str, np.ndarray]:
        """
        Reconstructs all vectors in the index and
        returns a dict: { doc_id: embedding_vector }.
        """
        total = self._faiss_index.ntotal
        dim = self._faiss_index.d

        try:
            raw = np.empty((total, dim), dtype="float32")
            self._faiss_index.reconstruct_n(0, total, raw)
        except AttributeError:
            raw = np.zeros((total, dim), dtype="float32")
            for internal_id in range(total):
                raw[internal_id] = self._faiss_index.reconstruct(internal_id)

        return { self._id_map[i]: raw[i] for i in range(total) }

if __name__ == "__main__":
    # 1) Load FAISS store
    faiss_store = FAISS.load_local(
        "../ExecutionModule/faiss",
        embeddings=embedding_model,
        allow_dangerous_deserialization=True
    )

    # 2) Extract all embeddings
    extractor = FaissEmbeddingExtractor(faiss_store)
    embeddings_dict = extractor.get_all_embeddings()

    # 3) Save in ./embedding/embeddings.pkl
    output_dir = "./embedding"
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, "embeddings.pkl")
    with open(output_path, "wb") as f:
        pickle.dump(embeddings_dict, f)

    print(f"Saved {len(embeddings_dict)} embeddings to {output_path}")
```
